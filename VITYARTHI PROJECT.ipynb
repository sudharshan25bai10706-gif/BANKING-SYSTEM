```python
print("\n===== WELCOME TO VIT BPL BANK ====\n")

import pandas as pd 
import math
import numpy as np
global df 
df=pd.DataFrame(columns=['Name','Pin','Min_deposit'])
global dep_df
dep_df=pd.DataFrame(columns=['Name','Dep_Date','Dep_amt'])
global with_df
with_df=pd.DataFrame(columns=['Name','Wth_Date','Wth_amt'])
global bal_df
bal_df=pd.DataFrame(columns=['Name','min_deposit','Dep_amt','Wth_amt','balance'])

def create_account():
    print("\n===== Create Account =====")
    name = input("Full name: ")
    pin=int(input("Enter a 4digit pin:"))
    pin_len=len(str(pin))
    if pin_len!=4:
        print("invalid pin")
    min_deposit=int(input("Enter the min balance:"))
    if min_deposit<500:
        print("Insufficient deposit")
    else:
        add_account(name,pin,min_deposit)
    print("SUCCESSFULLY ADDED")
    print(df.to_string(index=False))


def add_account(name,pin,min_deposit):
    global df
    df.loc[len(df)]=[name,pin,min_deposit]


def deposit():
    global dep_df
    print("===== The Deposited Amount =====")
    Name=input("Enter Account Name: ")
    Dep_Date=input(" Enter The Date (YYYY-MM-DD): ")
    Dep_amt=float(input("Enter The Deposit Amount: "))
    dep_df.loc[len(dep_df)]=[Name,Dep_Date,Dep_amt]
    print("SUCCESSFULLY DEPOSITED")
    print(dep_df.to_string(index=False))


def withdraw():
    global with_df
    print("===== The Withdrawn Amount =====")
    Name=input("Enter Account Name: ")
    Wth_Date=input(" Enter The Date (YYYY-MM-DD): ")
    Wth_amt=float(input("Enter The Withdraw Amount: "))
    with_df.loc[len(with_df)]=[Name,Wth_Date,Wth_amt]
    print("SUCCESSFULLY WITHDRAWN ")
    print(with_df.to_string(index=False))


def balance():
    print("===== Your Balance =====")
    Name = input("Enter Account Name: ")
    row = df.loc[df["Name"] == Name]
    if row.empty:
        print("Account not found.")
        return
    print(row.to_string(index=False))
    dep_row = dep_df.loc[dep_df["Name"] == Name]
    dep_sum = dep_row["Dep_amt"].sum()
    print("Total Deposited:", dep_sum)
    wth_row = with_df.loc[with_df["Name"] == Name]
    wth_sum = wth_row["Wth_amt"].sum()
    print("Total Withdrawn:", wth_sum)
    min_deposit = row.iloc[0]["Min_deposit"]
    balance_amt = min_deposit + dep_sum - wth_sum
    print("Current Balance:", balance_amt)


def list_accounts():
    print("=====All Bank Accounts=====")
    print(df.to_string(index=False))
    
    

def main_menu():
    while True:
        print("Welcome to vit bpl bank")
        print("1) Create account")
        print("2) Deposit")
        print("3) Withdraw")
        print("4) Check balance")
        print("5) List all accounts ")
        print("0) Exit")
        choice = input("Select an option: ")
        if choice == "1":
            create_account()
        elif choice == "2":
            deposit()
        elif choice == "3":
            withdraw()
        elif choice == "4":
            balance()
        elif choice == "5":
            list_accounts()
        elif choice == "0":
            print("Goodbye!")
            break
        else:
            print("Invalid option")
main_menu()
```
